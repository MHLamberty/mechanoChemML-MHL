```python
lam = 1.0398594822701848
mu  = 4.9802773854662625

nu = lam / (2*(lam + mu))
E  = mu * (3*lam + 2*mu) / (lam + mu)
K  = lam + 2*mu/3

print("nu =", nu)
print("E  =", E)
print("K  =", K)
```

```python
E_abaqus  =  10
nu_abaqus =  0.3

mu_expected  = E_abaqus/(2*(1+nu_abaqus))
lam_expected = E_abaqus*nu_abaqus/((1+nu_abaqus)*(1-2*nu_abaqus))

print("lambda expected =", lam_expected)
print("mu expected     =", mu_expected)
```

```python
lambda_vsi = 5.76923128947757
mu_vsi  = 3.846153664909679

E_abaqus  = 10
nu_abaqus = 0.3

mu_exp  = E_abaqus/(2*(1+nu_abaqus))
lambda_exp = E_abaqus*nu_abaqus/((1+nu_abaqus)*(1-2*nu_abaqus))

print("lambda expected =", lambda_exp)
print("mu expected     =", mu_exp)
print("\nlambda VSI      =", lambda_vsi)
print("mu VSI          =", mu_vsi)

print("\nlambda % error =", 100*(lambda_vsi-lambda_exp)/lambda_exp)
print("mu % error     =", 100*(mu_vsi-mu_exp)/mu_exp)
```
